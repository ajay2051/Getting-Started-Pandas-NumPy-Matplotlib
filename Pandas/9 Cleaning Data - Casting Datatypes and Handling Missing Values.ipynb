```python
import pandas as pd
import numpy as np
```

```python
people = {
    "first": ["john","dane", "jack", "chris", np.nan, None, 'NA'],
    "last": ["doe","martin","smith", "schafer", np.nan, np.nan, "Missing"],
    "email": ["johndoe@gmail.com", "danemartin@gmail.com","jacksmith@gmail.com", None, np.nan, np.nan, "Anonymous@email.com"],
    "age": ["33", "36", "63", "45", None, None, "Missing"]
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis="index", how="all")
```

```python
# drop rows if email is missing
# axis="index" means rows

df.dropna(axis="index", how="any", subset=["email"])
```

```python
df.replace("NA", np.nan, inplace=True)
df.replace("Missing", np.nan, inplace=True)
df
```

```python
df.isna()
```

```python
df.fillna("MISSING")
```

```python
df.fillna(0)
```

```python
df.dtypes
```

```python
# Change age to float

df["age"] = df["age"].astype(float)
```
