```python
from matplotlib import pyplot as plt
```

```python
minutes = [1,2,3,4,5,6,7,8,9]

player_1 = [1,2,2,3,3,3,4,4,4]
player_2 = [1,1,2,3,3,3,3,4,4]
player_3 = [1,1,2,3,3,3,3,4,4]

labels = ["Player_1", "Player_2", "Player_3"]
colors = ['blue','green','yellow']

plt.stackplot(minutes, player_1, player_2, player_3, labels=labels, colors=colors)

plt.legend(loc="upper left")

plt.title("Awesome Pie Chart")
plt.tight_layout()
plt.show()
```
