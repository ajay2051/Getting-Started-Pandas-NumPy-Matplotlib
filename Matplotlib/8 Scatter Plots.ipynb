```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
plt.style.use('seaborn')



x = [5,7,6,8,7,2,1,4,8,9,6,1,4]
y = [3,5,6,9,8,7,4,7,1,4,5,4,7]

colors = [4,5,6,8,9,7,8,7,1,7,5,3,7]

plt.scatter(x, y, s=100, c=colors, cmap='Greens', marker='X', edgecolor='black', linewidth=1, alpha=0.75)

cbar = plt.colorbar()
cbar.set_label('Satisfaction')


plt.show()
```

```python
data = pd.read_csv('scattered_plot_data.csv')
view_count = data['view_count']
likes = data['likes']
ratio = data['ratio']

plt.scatter(view_count, likes, edgecolor='black', c=ratio, cmap='summer', linewidth=1, alpha=0.75)


plt.title("Trending YouTube Videos")
plt.xlabel("View Count")
plt.ylabel("Total Likes")

plt.xscale('log')
plt.yscale('log')

cbar = plt.colorbar()
cbar.set_label('Like/Dislike Ratio')

plt.tight_layout()
plt.show()
```
